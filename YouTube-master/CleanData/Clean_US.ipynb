```python
# Dependencies
import pandas as pd
import numpy as np
import datetime as dt
import os,sys
from scipy import stats
```

```python
# Store filepath in a variable
us_file = "youtube_trending_US1.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
yt_us_df = pd.read_csv(us_file)
```

```python
yt_us_df.head()
```

```python
yt_us_df.shape
```

```python
yt_us_new_DF = yt_us_df
yt_us_new_DF.head()
```

```python
# Check how many rows have null values
yt_us_new_DF.isnull().sum()
```

```python
# Only keep rows where there is a valid video_id

yt_us_new_DF = yt_us_new_DF[yt_us_new_DF['video_id'].notna()]
yt_us_new_DF.isnull().sum()
```

```python
# Change data type of trending_date column

yt_us_new_DF['trending_date'] = yt_us_new_DF['trending_date'].astype('datetime64[ns]')
yt_us_new_DF.head()
```

```python
# Create new column for last day trending video appears in dataset
yt_us_new_DF['max_date'] = yt_us_new_DF.groupby('video_id').trending_date.transform('max')
yt_us_new_DF.head()
```

```python
# Create new column for first day trending video appears in dataset

yt_us_new_DF['min_date'] = yt_us_new_DF.groupby('video_id').trending_date.transform('min')
yt_us_new_DF.head()
```

```python
yt_us_new_DF.dtypes
```

```python
# Create new column for how many days video appears in trending dataset

yt_us_new_DF['days_trending'] = yt_us_new_DF["max_date"] - yt_us_new_DF["min_date"]
yt_us_new_DF.head()
```

```python
# Change datatype of days_trending column to string

yt_us_new_DF['days_trending'] = yt_us_new_DF['days_trending'].astype('str')
yt_us_new_DF.head()
```

```python
# Change days_trending column to integer

yt_us_new_DF['days_trending'] = yt_us_new_DF['days_trending'].str.rstrip(' days 00:00:00.000000000')
yt_us_new_DF.head()
```

```python
yt_us_new_DF['days_trending'] = yt_us_new_DF['days_trending'].str.strip()
```

```python
yt_us_new_DF['days_trending'].unique()
```

```python
yt_us_new_DF.isnull().sum()
```

```python
yt_us_new_DF['days_trending'] = yt_us_new_DF['days_trending'].str.replace('', '0')
```

```python
yt_us_new_DF['days_trending'] = '0' + yt_us_new_DF['days_trending'].astype(str)
```

```python
yt_us_new_DF['days_trending'].unique()
```

```python
yt_us_new_DF['days_trending'] = yt_us_new_DF['days_trending'].astype('int') +1
yt_us_new_DF['days_trending']
```

```python
# Get max views for rows that are latest date

yt_us_new_DF["max_views"] = yt_us_new_DF.query('trending_date==max_date')['views']                                           
yt_us_new_DF.head()
```

```python
# Get min views for rows that are the earliest date

yt_us_new_DF["min_views"] = yt_us_new_DF.query('trending_date==min_date')['views']                                           
yt_us_new_DF.head()
```

```python
# Get max likes & min likes for rows that are latest and earliest date

yt_us_new_DF["max_likes"] = yt_us_new_DF.query('trending_date==max_date')['likes']
yt_us_new_DF["min_likes"] = yt_us_new_DF.query('trending_date==min_date')['likes']        
yt_us_new_DF.head()
```

```python
# Get max dislikes and min dislikes for rows that are latest and earliest date

yt_us_new_DF["max_dislikes"] = yt_us_new_DF.query('trending_date==max_date')['dislikes']
yt_us_new_DF["min_dislikes"] = yt_us_new_DF.query('trending_date==min_date')['dislikes']        
yt_us_new_DF.head()
```

```python
# Get max dislikes and min dislikes for rows that are latest and earliest date

yt_us_new_DF["max_comments"] = yt_us_new_DF.query('trending_date==max_date')['comment_count']
yt_us_new_DF["min_comments"] = yt_us_new_DF.query('trending_date==min_date')['comment_count']        
yt_us_new_DF.head()
```

```python
yt_us_new_DF.dtypes
```

```python
yt_us_new_DF.to_csv('CleanData_US.csv', index=False, header=True)
```
